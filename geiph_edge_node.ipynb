```python
# -*- coding: utf-8 -*-
import jieba
import sys

from sklearn import feature_extraction
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = []
with codecs.open('C:/Users/orozcohsu/text source/20170424_120002.txt', 'r',encoding='utf8') as f:
    for line in f:
        corpus.append(" ".join(jieba.cut(line.split(',')[0], cut_all=False)))

#去除英文網址
def isEnglish(s):
    try:
        s.encode(encoding='utf-8').decode('ascii')
    except UnicodeDecodeError:
        return False
    else:
        return True


p=0

vectorizer = TfidfVectorizer()
tfidf = vectorizer.fit_transform(corpus)
print(tfidf.shape)

words = vectorizer.get_feature_names()
word=[]
similarity=[]

str=""
fk = open("C:/Users/orozcohsu/text source/20170424_120002_node.tmp.csv",'w', encoding='utf-8')

for i in range(len(corpus)):
    print('----Document %d----' % (i))
    for j in range(len(words)):
        if tfidf[i,j] > 0.05:
            if isEnglish(words[j]) == False:
                s = '%.4f' % tfidf[i,j]
                fk.write('%s,%s,%s\n' % (p, words[j], s))
                word.append(words[j])
    del str
    fp = open("C:/Users/orozcohsu/text source/20170424_120002_" + str(p) + ".txt",'w', encoding='utf-8')
    #print(word)
    str = ' '.join(word)
    similarity.append(str)
    fp.write(str + "\n")
    fp.close()
    p=p+1
    word=[]

fk.close()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vect = TfidfVectorizer(min_df=1)
tfidf = vect.fit_transform(similarity)
result = (tfidf * tfidf.T).A
```

```python
xx,yy=result.shape

fp = open("C:/Users/orozcohsu/text source/edge.tmp.csv",'w', encoding='utf-8')

node=[]
del str

#Geiph edge.tmp.csv
for x in range(0,xx):
    for y in range(0,yy):
        if result[x,y] < 0.99:
            if result[x,y] > 0.04 and x != y:
                print(str(x) + ":" + str(y) + ":" + str(result[x,y]))
                fp.write(str(x) + "," + str(y) + "\n")
                node.append(str(x))
            
fp.close()
```

```python
#Geiph edge.csv
import numpy as np

fp = open("C:/Users/orozcohsu/text source/edge.csv",'w', encoding='utf-8')
switch_pair=[]

#delete duplicates
with codecs.open('C:/Users/orozcohsu/text source/edge.tmp.csv', 'r',encoding='utf8') as f:
    for line in f:
        a,b = line.split(',')
        k = str(a.strip() + "," + b.strip())
        s = str(b.strip()) + "," + str(a.strip())
        switch_pair.append(s)
        if k in switch_pair:
            #print(line.strip())
            fp.write(line.strip() + "\n")

fp.close()
```

```python
article=0
token=[]
score=[]

#Geiph node.csv
fp = open("C:/Users/orozcohsu/text source/node.csv",'w', encoding='utf-8')

with codecs.open('C:/Users/orozcohsu/text source/20170424_120002_node.tmp.csv', 'r',encoding='utf8') as f:
    for line in f:
        a,b,c = line.split(',')
        if article == int(a):
            token.append('%s,%s' % (b.strip(),c.strip()))
            score.append(float(c))
        else:
            #print(token)
            m = max(score)
            #print("max:" + str(m))
            for x in token:
                #print("x:" + x)
                mm = x.split(',')[1]
                if float(m) == float(mm):
                    #only match the first highest score token as node label
                    #print("match max:" + x.split(',')[0])
                    fp.write('%s,%s\n' % (article,x.split(',')[0]))
                    break
            article = article+1
            del token[:]
            del score[:]
            token.append('%s,%s' % (b.strip(),c.strip()))
            score.append(float(c))

fp.close()
```
